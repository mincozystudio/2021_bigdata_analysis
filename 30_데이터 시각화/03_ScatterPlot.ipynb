# 3. Scatter Plot
* 좌표상의 점들을 표시하여 두 변수 간의 관계를 나타내는 그래프

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50) # [0, 1]
y = np.random.rand(50)
plt.scatter(x, y)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, y)

x2 = np.random.rand(50)
y2 = np.random.rand(50)
plt.scatter(x2, y2)
plt.show()
```

### Marker 지정하기
plt.scatter(x, y, color='r')
plt.show()

```python
# 점마다 다른 color 지정하기
x = [1, 2, 3]
y = [1, 2, 3]
colors = ['r', 'g', 'b']
plt.scatter(x, y, c=colors)
```

```python
# ColorMap(색상지도) 사용하기
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, c=x, cmap = 'Blues')
plt.colorbar()
plt.show()
```

```python
# Marker Size 변경하기
size = np.random.randint(50) * 10000
plt.scatter(x, y, c=x, cmap='Blues')
plt.show()
```

```python
# 투명도 ( alpha값, 0~1 )
plt.scatter(x, y, c=x, cmap='Blues', s = size, alpha = 0.5)
plt.show()
```
